```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas
from pandas.plotting import scatter_matrix

#read data
data = pandas.read_csv('costOfLivingIndex.csv')
column_list=data.columns
```

```python
print(column_list)
```

```python
#specify columns that show relation 1-COLI
scatter_matrix(data[['refugees', 'costOfLivingIndex', 'rentIndex', 'costOfLivingPlusRentIndex', 'groceriesIndex', 'restaurantPriceIndex', 'localPurchasingPowerIndex']])
plt.show()
```

```python
#specify columns that show relation 2-COLI
scatter_matrix(data[['refugees', 'costOfLivingIndex', 'rentIndex']])
plt.show()
```

```python
#specify columns that show relation 3-COLI
scatter_matrix(data[['refugees','costOfLivingPlusRentIndex', 'groceriesIndex']])
plt.show()
```

```python
#specify columns that show relation 4-COLI
scatter_matrix(data[['refugees','restaurantPriceIndex', 'localPurchasingPowerIndex']])
plt.show()
```
